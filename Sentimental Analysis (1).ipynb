Here we have printed the stopwords, but we have removed the word "not" as it can define a different meaning to a word

```python
%%time
l = []
for i in rawdata['Text']:
    word = []
    sent = i.lower()
    sent = emoji.demojize(sent)
    sent = re.sub('â€˜|â€™',"'",sent)
    sent = re.sub('&amp;',"and",sent)
    sent = re.sub('<.*?>','',sent)
    sent = re.sub('https?://\S*|www\S*','',sent)
    sent = re.sub(' @\w+| #\w+',"",sent)
    sent = re.sub("[^a-z0-9 ]","",sent)
    for i in word_tokenize(sent):
        if i in words:
            pass
        else:word.append(PorterStemmer().stem(i))
    l.append(" ".join(word))
```

The above step is PreProcessing where we clean the data

```python
cleaned_data=pd.DataFrame({"cleaned_Text":l})
```

```python
cleaned_data
```

BOW- Bag Of Words

```python
tf=CountVectorizer()
transform=tf.fit_transform(cleaned_data["cleaned_Text"][0:25000])
```
